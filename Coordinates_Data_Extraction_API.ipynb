```python
import requests
import numpy as np
from config import g_key,weather_api_key
import pandas as pd
from pprint import pprint
from json.decoder import JSONDecodeError
import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv("Resources/1-raw_vineyard_data/winemag-data-130k-v2.csv",index_col=0)
data
```

```python
wine_data = data.dropna(subset = ['price','province','region_1','region_2']).reset_index(drop=True)
wine_data=wine_data.drop("winery",axis=1)
wine_data
```

```python
wine_data.to_csv("filtered_wine_data.csv")
```

```python
df = wine_data.drop_duplicates(subset=['province','region_1','region_2','winery_name'])
df
```

```python
new_data = wine_data[["region_1","region_2","winery_name"]]
new_data
```

```python
new_data['Location1']=new_data['winery_name']+' ' + new_data['region_1']
new_data['Location2']=new_data['winery_name']+' '+ new_data['region_2']
new_data
```

```python
new_data= new_data.drop_duplicates(['Location1','Location2'])
new_data
```

```python
new_data['winery_name'].value_counts()
```

```python
new_data["Latitude"]=""
new_data["Longitude"]=""
new_data.drop(["region_1","region_2"],axis=1)
# new_data=new_data.head(50)
new_data
```

```python
for index,row in new_data.iterrows():
    zeroResultCount=0
    name = row["Location1"]
    params = { "key": g_key,
               "input":name,
              "inputtype":"textquery",
             "fields": "geometry"}
    url = "https://maps.googleapis.com/maps/api/place/findplacefromtext/json?"
#     print(url+"::"+params['input'])
    try:
        response = requests.get(url,params=params).json()
        
        if(response['status']== "ZERO_RESULTS"):
                zeroResultCount+=1
                name = row["Location2"]
                params = { "key": g_key,
               "input":name,
              "inputtype":"textquery",
             "fields": "geometry"}                  
        response = requests.get(url,params=params).json()
        new_data.loc[index,"Latitude"] = response["candidates"][0]["geometry"]["location"]["lat"]
        new_data.loc[index,"Longitude"] = response["candidates"][0]["geometry"]["location"]["lng"]
    except(KeyError, IndexError, JSONDecodeError):
        print(f"{index} {name} not found. Skipping...")
    except requests.ConnectionError:
        print("ConnectionError...")
    except requests.Timeout:
        print("Request Timeout...")
print(zeroResultCount)       
```

```python
new_data['Latitude'].replace('', np.nan, inplace=True)
new_data['Longitude'].replace('', np.nan, inplace=True)
new_data
```

```python
df=new_data.dropna(subset = ["Latitude","Longitude"]).reset_index(drop=True)
df
```

```python
df.to_csv("locations.csv")
```

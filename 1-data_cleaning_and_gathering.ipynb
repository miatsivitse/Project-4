```python
import requests
from config import g_key
import pandas as pd
from pprint import pprint
from json.decoder import JSONDecodeError
import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv("Resources/1-raw_vineyard_data/winemag-data-130k-v2.csv",index_col=0)
data
```

```python
wine_data = data.dropna(subset = ['price','province','region_1','region_2']).reset_index(drop=True)
# for index,row in wine_data.iterrows():
#      wine_data.loc[index,"winery"]=row["winery"] + " winery"
wine_data=wine_data.drop("winery",axis=1)
# wine_data=wine_data.head(100)
wine_data
```

```python
# create new new columns for lat and lng
wine_data["Latitude"]=""
wine_data["Longitude"]=""
wine_data
```

```python
# https://maps.googleapis.com/maps/api/place/findplacefromtext/json?input=Marthas%20vineyard&inputtype=textquery&fields=geometry&key=API_KEY
name=""
# loop through the df 
zeroResultCount=0
for index,row in wine_data.iterrows():
    name= row["winery_name"] +" "+row["region_1"]
    params = { "key": g_key,
               "input":name,
              "inputtype":"textquery",
             "fields": "geometry"}
    url = "https://maps.googleapis.com/maps/api/place/findplacefromtext/json?"
#     print(url+"::"+params['input'])
    try:
        response = requests.get(url,params=params).json()
        
        if(response['status']== "ZERO_RESULTS"):
                zeroResultCount+=1
                name1= row["winery_name"] +" "+row["region_2"]
                params = { "key": g_key,
               "input":name1,
              "inputtype":"textquery",
             "fields": "geometry"}                  
        response = requests.get(url,params=params).json()
        wine_data.loc[index,"Latitude"] = response["candidates"][0]["geometry"]["location"]["lat"]
        wine_data.loc[index,"Longitude"] = response["candidates"][0]["geometry"]["location"]["lng"]
    except(KeyError, IndexError, JSONDecodeError):
        print(f"{index} {name} not found. Skipping...")
    except requests.ConnectionError:
        print("ConnectionError...")
    except requests.Timeout:
        print("Request Timeout...")
print(zeroResultCount)       
```
